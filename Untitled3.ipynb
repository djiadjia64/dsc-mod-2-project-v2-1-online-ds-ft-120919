Now, I will use feature engineering and combine the related dataframes. 

```python
loglstat = kc['sqfttotal']
kc['scaled_sqft']=(loglstat-np.mean(loglstat))/(max(loglstat)-min(loglstat))
kc['scaled_sqft1']=(loglstat-np.mean(loglstat))/np.sqrt(np.var(loglstat))
```

```python
one = kc.sqft_above
kc['above1'] = (one-np.mean(one))/np.sqrt(np.var(one))
two = kc.sqft_living
kc['living1'] = (two-np.mean(two))/np.sqrt(np.var(two))
three = kc.sqft_living15
kc['living15'] = (three-np.mean(three))/np.sqrt(np.var(three))
```

```python
kc['total'] = (kc['above1'] + kc['living1'] + kc['living15'])/3
```

```python
sns.distplot(kc.total)
```

```python
kc.total
```

Now repeating these steps to combine grade and bathrooms

```python
four = kc.bathrooms
five = kc.grade
kc['bathrooms1']=(four-np.mean(four))/np.sqrt(np.var(four))
kc['grade1']=(five-np.mean(five))/np.sqrt(np.var(five))
#standardizing with this method
```

```python
kc['total2'] = (kc['bathrooms1'] + kc['grade1'])/2
```

```python
kc.total2.describe()
```

```python
sns.distplot(kc.total2)
```

```python
six = kc.price
kc['price1'] = (six-np.mean(six))/np.sqrt(np.var(six))
sns.distplot(kc.price1)
```

```python
import statsmodels.api as sm

y = pd.DataFrame(kc, columns = ['price1'])
X = pd.concat([kc.total, kc.total2], axis=1)

X_int = sm.add_constant(X)
model = sm.OLS(y, X_int.astype(float)).fit()

model.summary()
#conditional number greatly reduced by we did lose some predictive value in our R-squared
```
